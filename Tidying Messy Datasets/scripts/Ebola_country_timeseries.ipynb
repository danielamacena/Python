```python
from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials

gc = gspread.authorize(GoogleCredentials.get_application_default())
spreadsheet = gc.open_by_url('https://docs.google.com/spreadsheets/d/1MS-RvEkO-YyLZoDQq4vzQHHg19OH30OQupSqrKoY9pk/edit?usp=sharing')

import pandas as pd
page = spreadsheet.sheet1
data = pd.DataFrame(page.get_all_records())
```

```python
ebola_melt = pd.melt(data,
                     id_vars=['Date', 'Day'],
                     var_name='cd_country',
                     value_name='count')
ebola_melt
```

```python
var_split = ebola_melt['cd_country'].str.split('_')
```

```python
status_values = var_split.str.get(0)
country_values = var_split.str.get(1)
```

```python
ebola_melt['status'] = status_values
ebola_melt['country'] = country_values
```

```python
variable_split = ebola_melt['cd_country'].str.split('_', expand=True)
variable_split
```

```python
variable_split.columns = ['status1', 'country1']
variable_split.head()
```

```python
ebola_clean = pd.concat([ebola_melt, variable_split], axis=1)
ebola_clean
```
