```python
from google.colab import auth
auth.authenticate_user()

import gspread
from oauth2client.client import GoogleCredentials

gc = gspread.authorize(GoogleCredentials.get_application_default())
spreadsheet = gc.open_by_url('https://docs.google.com/spreadsheets/d/1vxUtdJeCsR6i_H-X2kocIPO3OK-M46LPMLrBdFx2Noo/edit?usp=sharing')
```

```python
import pandas as pd
page = spreadsheet.sheet1
data = pd.DataFrame(page.get_all_records())
data
```

```python
dataset_religions=pd.melt(data, id_vars='religion',
        var_name='income', value_name='count')
```
