````python
tb = pd.read_csv(io.BytesIO(uploaded['tb.csv']))
tb.head()
````

````python
ebola = pd.read_csv(io.BytesIO(uploaded['ebola_country_timeseries.csv']))
````

````python
ebola.head()
````

````python
# first let's melt the data down
ebola_long = ebola.melt(id_vars=['Date', 'Day'],
                        value_name='count',
                        var_name='cd_country')
ebola_long.head()


````

````python
var_split = ebola_long['cd_country'].str.split('_')
var_split

````

````python
type(var_split)
````

````python
var_split[0]
````

````python
var_split[0][0]
````

````python
var_split[0][1]
````

````python
# save each part to a separate variable
status_values = var_split.str.get(0)
country_values = var_split.str.get(1)
````

````python
status_values.head()
````

````python
country_values.head()
````

````python
ebola_long
````

````python
# assign the parts to new dataframe columns
ebola_long['status'] = status_values
ebola_long['country'] = country_values
````

````python
ebola_long.head()
````

### above in a single step

````python
variable_split = ebola_long['cd_country'].str.split('_', expand=True)
variable_split.head()
````

````python
variable_split.columns = ['status1', 'country1']
````

````python
ebola = pd.concat([ebola_long, variable_split], axis=1)
ebola.head()
````

## Variables in both rows and columns

````python
weather = pd.read_csv(io.BytesIO(uploaded['weather.csv']))
````

````python
weather.head()
````

````python
weather_melt = pd.melt(weather,
                      id_vars=['id', 'year', 'month', 'element'],
                      var_name='day',
                      value_name='temp')
````

````python
weather_melt.head()
````

````python
weather_tidy = weather_melt.pivot_table(
    index=['id', 'year', 'month', 'day'],
    columns='element',
    values='temp')
weather_tidy.head()
````

````python
weather_flat = weather_tidy.reset_index()
weather_flat.head()
````
